### matplotlib 사용

```python
import matplotlib.pyplot as plt

y = [1, 2, 3, 4, 5, 6]
plt.plot(y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [4, 7, 6, 8, 2, 9]

plt.plot(x, y, "ro")
plt.show()
```

```python
# 여러 차트 그리기 (비추천)

x = [0, 1, 2, 3, 4, 5]
y = [4, 7, 6, 8, 2, 9]
y2 = [6, 9, 12, 18, 22, 14]
y3 = [9, 22, 3, 16, 10, 15]

plt.plot(x, y, "r--", x, y2, "bs", x, y3, "g^")       # y값 차트 출력 (비추천)

plt.show()
```

```python
# 여러 차트 그리기 (비추천)

x = [0, 1, 2, 3, 4, 5]
y = [4, 7, 6, 8, 2, 9]
y2 = [6, 9, 12, 18, 22, 14]
y3 = [9, 22, 3, 16, 10, 15]

plt.plot(x, y, "r--", x, y2, "b-.", x, y3, "g:")       # y값 차트 출력 (비추천)

plt.show()
```

```python
# 여러 차트 그리기 (추천)

y = [4, 7, 6, 8, 2, 9]
y2 = [6, 9, 12, 18, 22]

plt.plot(y)
plt.plot(y2)

plt.show()
```

```python
# 여러 차트 그리기

x = [0, 1, 2, 3, 4, 5]
y = [4, 7, 6, 8, 2, 9]
y2 = [6, 9, 12, 18, 22, 14]
y3 = [9, 22, 3, 16, 10, 15]

plt.plot(x, y3, "r", label="y3-data")
plt.plot(x, y, "g", marker="^", label="y-data")

plt.legend()   # 범례 표시
plt.title("Graph Title", size=16)

plt.show()
```
