#### for 한줄쓰기
- [결과 for 반복문]



```python
# "연습" 이란 단어를 5번 생성하는 리스트 구조 변수 만들기

for i in range(5):
    print("연습")

# 한줄 쓰기

["연습" for i in range(5)]
```

```python
tmp1 = []

for i in range(5):
    tmp1.append("연습")

tmp2 = ["연습" for i in range(5)]

print(tmp1)
print(tmp2)
```

#### 이중 for 한줄쓰기
- [결과 for 반복문1 for 반복문2]

```python
val = [val, [11, 12, 13]]

print(val)
```

```python
val_tmp1 = []
for i in val:
    for j in i:
        val_tmp1.append(j)

print(val_tmp1)

val_tmp2 = [j for i in val for j in i]

print(val_tmp2)
```

#### if가 포함된 for 한줄쓰기
- [결과 for 반복문 if 조건문]

```python
val = list(range(1, 11))

print(val)

# 홀수 데이터만 추출하여 리스트로 출력하기

odd1 = []
for i in val:
    if i%2 == 1:        # a % b = a를 b로 나누고 남은 나머지(나머지값 계산)
        odd1.append(i)

print(f"홀수 데이터(기본): {odd1}")

# 한줄쓰기로 표현

odd2 = [i for i in val if i%2==1]

print(f"홀수 데이터(한줄쓰기)): {odd2}")
```

#### for문에 if ~ else ~ 구조 한줄쓰기
- [T 결과 if 조건문 else F 결과 for 반복문]

```python
# 리스트 변수에 val의 짝수 / 홀수 표기

tmp1 = []
for i in val_tmp1:
    if i%2 == 1:
        tmp1.append("홀수")
    else:
        tmp1.append("짝수")

print(tmp1)

# 한줄쓰기

tmp2 = ["홀수" if i%2==1 else "짝수" for i in val_tmp1]

print(tmp2)
```

### 미션
- val1 <= 1 ~ 10 까지 정수형 리스트 생성
- val2 <= 11 ~ 20 까지 홀수값만 가지는 리스트 생성
- val = [val1, val2]
- tmp1 일반 for문 [짝수/홀수]
- tmp2 한줄쓰기 [짝수/홀수]

```python
val = [list(range(1, 11)), list(range(11, 20, 2))]
#print(val)

tmp1 = []
for i in val:
    for j in i:
        if j%2 == 1:
            tmp1.append("홀수")
        else:
            tmp1.append("짝수")
            
print(tmp1)
```

```python
tmp2 = ["홀수" if j%2==1 else "짝수" for i in val for j in i]

print(tmp2)
```

### 에러처리
- try: 정상~ except: 에러처리

```python
print(aaa)  # aaa 변수가 없기 때문에 에러 발생
```

```python
## 기본적인 형식
try:
    print(aaa)
except:
    print("에러발생")
```

```python
## try: ~ except: ~ 주의점
try:
    print("안녕하세요.")  # 정상 코드
    print(aaa)            # 에러 발생 코드
    print("반갑습니다.")  # 정상 코드
except:
    print("에러발생")
```

```python
## 기본적인 형식/pass
try:
    print(aaa)
except:
    pass   # 넘어가시오

print("에러 무시")
```

```python
## 에러 발생에 대한 원인을 메세지로 전달(즉, 코드 실행시 멈추지 않고 계속 실행)
try:
    print(aaa)
except Exception as e:
    print(f"에러발생원인:{e}") 

print("=== END ===")
```

#### random 모듈
- 파이썬에서 난수를 발생시키는 기본 모듈
- 참조 : https://docs.python.org/ko/3/library/random.html
- random 함수
 * random() : 0 ~ 1 사이의 숫자를 임의로 출력
 * randrange(시작, 종료) : 시작부터 종료값 사이의 숫자를 임의로 발생
 * randint(a, b) : a와 b사이의 정수형 숫자 발생
 * shuffle() : 스퀀스 데이터의 순서를 임의로 변경해서 출력
 * choice() : 스퀀스 데이터들 중 하나를 뽑아 출력

```python
import random

random.random()
```

```python
# a에서 b 사이의 임의값 출력

print(random.randint(1, 46))

print(random.randrange(1, 101, 2))   # step 값 지정 가능
```

```python
# 스퀀스 데이터 처리

tmp = ["a", "b", "c", "d", "e"]

random.shuffle(tmp)      # 순서 변경하기

print(tmp)

print(random.choice(tmp))

메뉴 = ["라면", "김밥", "자장면", "짬뽕"]

print(random.choice(메뉴))
```

```python
num = list(range(1, 46))

# print(num)

random.sample(num, k=7)
```

### 미션 [숫자 맞추기]
- 1 ~ 100사이의 임의값 생성 후 tmp 저장
- 사용자로부터 1 ~ 100 사이값을 입력 받기
- 최초 생성된 숫자와 일치할때까지 입력(while / for)
  * 입력값이 tmp 보다 크면 "다운" 출력
  * 입력값이 tmp 보다 작으면 "업" 출력
  * 일치하면 "축하합니다. 00회 만에 맞췄습니다." 출력
  * 정답을 맞춘 횟수 계산
- 만약 문자를 입력시 오류 메시지 출력 후 다시 입력
  * "에러 발생 : 숫자 데이터만 입력하세요."

```python
## 반복문(while)/break
import random
number = random.randint(1, 100)
#print(number)
game_count = 1

while True:
    try:
        my_number = int(input("1~100 사이의 숫자를 입력하세요:"))
        
        if my_number > number:
            print("다운")
        elif my_number < number:
            print("업")
        elif my_number == number:
            print(f"축하합니다.{game_count}회 만에 맞췄습니다")
            break
        
        game_count = game_count + 1
    except:
        print("에러가 발생하였습니다. 숫자를 입력하세요")
```
