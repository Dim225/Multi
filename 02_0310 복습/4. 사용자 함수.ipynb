```python
# 실행문만 있는 함수

# 함수 생성
def prn_hello(name, age):
    print(f'{name}님의 나이는 {age}입니다.')
    print("함수 실행 종료")

# 함수 호출
prn_hello("파이썬", "10")
```

```python
# 두 수를 입력받아 처리하는 함수

def add(a, b):          # a변수, b변수 => 지역변수
    tot = a + b
    return tot

# 함수 호출
tot = 0
ptot = add(10, 20)
print(f"함수 호출 결과 : {ptot}")
#print(f"함수a:{a}, 함수b:{b}")

###함수에서 여러개의 결과값 전달###
def add1(a, b):          # a변수, b변수 => 지역변수
    tot = a + b
    return a, b, tot

pa, pb, ptot = add1(10, 20)
print(f"함수 호출 결과 : {ptot}")
print(f"함수a:{pa}, 함수b:{pb}")
```
