### 2. 학습과 테스트 데이터로 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, random_state=2022            # train : test = 3 : 1
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
np.unique(y_train, return_counts=True)
```

```python
# y값을 균등하게 뽑으려면 stratify 옵션을 설정
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, stratify=iris.target, random_state=2022
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
np.unique(y_train, return_counts=True)
```

```python
# test size 비율을 20%로 설정
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, stratify=iris.target, test_size=0.2, random_state=2022
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
np.unique(y_train, return_counts=True)
```

### 3. 학습

- 결정트리(Decision Tree)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# 모델 생성 - 객체 생성
dtc = DecisionTreeClassifier(random_state=2022)
```

```python
# 모델 하이퍼 파라메터
dtc.get_params()
```

```python
# 학습(훈련)
dtc.fit(X_train, y_train)
```

### 4. 예측

```python
pred_dt = dtc.predict(X_test)
```

```python
res = pd.DataFrame({'y':y_test, 'DT':pred_dt})
res.head()
```

### 5. 평가

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred_dt)
```

```python
# 4, 5번을 한꺼번에 수행
dtc.score(X_test, y_test)
```

- Support Vector Machine(SVM)

```python
# 3) 모델 생성 및 학습
from sklearn.svm import SVC
svc = SVC(random_state=2022)
svc.fit(X_train, y_train)
```

```python
# 4) 예측
pred_sv = svc.predict(X_test)
```

```python
# 5) 평가
accuracy_score(y_test, pred_sv)
```

- Logistic Regression

```python
# 3) 모델 생성 및 학습
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=2022, max_iter=500)
lrc.fit(X_train, y_train)
```

```python
# 4) 예측
pred_lr = lrc.predict(X_test)
```

```python
# 5) 평가
accuracy_score(y_test, pred_lr)
```

- max_iter 기본 값 오류 해결 방안

```python
lrc2 = LogisticRegression(random_state=2022)
lrc2.fit(X_train, y_train)                      # 기본 max_iter 100이지만 100번 학습은 부족하여 오류 발생
```
