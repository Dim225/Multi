### ndarray 생성

- list, tuple 등의 시퀀스 자료형으로 구성현할 수 있으며 내부 요소의 자료형은 모두 동일한 타입으로만 가질수 있다.
- 자료의 크기도 고정되어 있음

```python
arr = np.array((1, 2, 3))
print(arr)
print(arr.dtype)
print(arr.shape)
```

```python
arr = np.array(((1, 2, 3.), (1, 2, 3.)))
print(arr)
print(arr.dtype)
print(arr.shape)
```

```python
arr = np.array([[(1, 2, 3.), (10, 20, "30")],
                [(1, 33, 25), (15., 22.0, 35)],
                [(10, 3, 5), (15., 22.0, 35)]])
print(arr)
print(arr.dtype)
print(arr.shape)  # 1차원:(m, ) / 2차원:(m, n) / 3차원:(p, q, r)
```

```python
arr = np.array((1, 2, 3), dtype='float64')
print(arr)
print(arr.dtype)
print(arr.shape)
```

```python
# 새로운 데이터를 받아와 순차적으로 array 만들기
arr = np.array([])

for i in range(3):
    arr = np.append(arr,[10, 20, 30])

print(arr)
```

```python
# np.arange 순차적인 배열값 만들기
arr = np.arange(15)
arr
```

```python
# reshape로 차원 조정하기
arr = np.arange(18).reshape(3, 6)   # reshape(3, 5)의 숫자의 곱은 array(15)의 숫자와 일치
print(arr, "\n")

arr1=arr.reshape(2, 9)    # (3, 6)의 구조를 (2, 9)로 변경
print(arr1, "\n")

arr1=arr.reshape(2, -1)   # 행을 기준으로 열 원소를 자동으로 계산
print(arr1, "\n")

arr1=arr.reshape(-1, 9)    # 원소의 갯수를 기준으로 행의 갯수 자동 계산
print(arr1, "\n")

arr1=arr.reshape(-1)      # 1차원 구조로 변경
print(arr1, "\n")

arr1=arr.reshape(1, -1)      # 1차원 구조로 변경
print(arr1, "\n")
```

### array 연산

```python
# 함수를 이용한 계산(합계)
print(arr)
print("전체 데이터 합계:", np.sum(arr)) 
print("열 기준 합계/axis=0:", np.sum(arr, axis=0)) 
print("행 기준 합계/axis=1:", np.sum(arr, axis=1)) 
```

```python
# array간 사칙연산
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

# 더하기
print("더하기:",arr1 + arr2)

# 빼기
print("빼기",arr1 - arr2)

# 곱하기
print("곱하기",arr1 * arr2)

# 나누기
print("나누기", arr2 + arr1)
```

```python
# array간 사칙연산
arr1 = np.array([1, 2, 3, 4]).reshape(2,-1)
arr2 = np.array([5, 6, 7, 8]).reshape(2,-1)

# 더하기
print("더하기:",arr1 + arr2)

# 빼기
print("빼기",arr1 - arr2)

# 곱하기
print("곱하기",arr1 * arr2)

# 나누기
print("나누기", arr2 + arr1)
```

```python
# array간 사칙연산
arr1 = np.array([1, 2, 3, 4]).reshape(2,-1)
arr2 = np.array([5, 6])

# 더하기
print("더하기:",arr1 + arr2)

# 빼기
print("빼기",arr1 - arr2)

# 곱하기
print("곱하기",arr1 * arr2)

# 나누기
print("나누기", arr2 + arr1)
```

### 배열에 대한 인덱싱/합치기

```python
arr=np.arange(0, 20, 2)
print(arr)
```

```python
# 요소 인덱싱
print(arr[0])
print(arr[[0, 3, 5]])
print(arr[2:5])
```

```python
## True/False로 구성된 boolean 배열 기준 출력
print(arr >= 10)
print(arr[arr >= 10])
```

```python
## 0이 아닌 값만 출력
arr[[3, -1]]=0
print(arr)
print(np.nonzero(arr))
```

```python
arr=arr.reshape(2, -1)
print(arr, "\n")
print(arr[1, 1:4])  # 행 1이면서 열이 1~4 사이값 출력
print(arr[:, 1:4])  # 모든 행에서 열이 1~4 사이값 출력
```

```python
# array 합치기

arr1 = np.arange(1, 5).reshape(2, -1)
arr2 = np.array([5, 6]).reshape(1, 2)   # arr2 = np.array([[5, 6]])

print(arr1.ndim, arr2.ndim)

print(arr1, "\n", arr2, "\n")
print(np.concatenate((arr1, arr2), axis=0), "\n")
print(np.concatenate((arr1, arr2.T), axis=1), "\n")
print(np.concatenate((arr1, arr2.reshape(2,-1)), axis=1), "\n")
```

#### numpy 입출력
- np.save(file, arr)  => array 1개를 파일로 저장(바이너리)
- np.savez(file, arr1, arr2) => 복수 배열을 파일로 저장(바이너리)
- np.load(file)   => 배열 형식으로 저장된 데이터 불러오기
- np.savetxt()   => 텍스트 형식으로 배열값 저장
- np.loadtxt()   => 텍스트 형식으로 저장된 배열값 불러오기

```python
def pprint(arr):
    print(f'type: {type(arr)}')
    print(f'shape:{arr.shape}, ndim:{arr.ndim}, dtype:{arr.dtype}')
    print(f'array data: \n{arr}')
```

```python
# np.random.randint(시작값, 종료값, 갯수)
x = np.random.randint(0, 10, (2, 3))
y = np.random.randint(0, 10, (2, 3))

pprint(x)
pprint(y)
```

```python
# 1개의 배열 저장
np.save("./data/arr1", x)
```

```python
# x, y 두개의 배열 저장
np.savez("./data/arr_all", x, y)
```

```python
# .npy 파일 읽어오기
npy_file=np.load("./data/arr1.npy")

pprint(npy_file)
```

```python
# .npz 파일 읽어오기
npz_file=np.load("./data/arr_all.npz")

print(npz_file.files)

pprint(npz_file['arr_0'])
pprint(npz_file['arr_1'])
```

### 이미지 가져와 배열 형식으로 저장/읽어오기

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 이미지 폴더 정의
fdirs=['ants', 'bees']

# 이미지에 대한 픽셀정보(X)와 레이블정보(Y) 저장
X_pix = []
y_lab = []

# 이미지 읽어와 배열로 저장하기

for inx, fdir  in enumerate(fdirs):
    files = os.listdir(fdir)

    for f in files:
        # 이미지 읽어오기        
        img = cv2.imread(fdir + "/" + f, 0)
        # 이미지 사이트 변경하기
        img = cv2.resize(img, (64, 64))
        # 이미지 ndarray로 변경하기
        img = np.array(img).reshape(-1).tolist()
        # X 리스트에 이미지 데이터 추가
        X_pix.append(img)

        # y 레이블 데이터 추가(0 = ants / 1 = bees)
        y_lab.append(inx)
```

```python
len(X_pix)
```

```python
# 저장한 이미지 1개 꺼내와 출력하기
plt.imshow(np.array(X_pix[75]).reshape(64, 64), cmap='gray')
plt.show()
```

배열 이미지 저장하기

```python
# 이미지 정보를 배열 구조로 변경하고 데이터 구조 확인
X_pix = np.array(X_pix)
y_lab = np.array(y_lab)

pprint(X_pix)
pprint(y_lab)

# 이미지 픽셀 정보와 레이블 정보를 하나로 만들기
Xy = np.concatenate((X_pix, y_lab.reshape(157, -1)), axis=1)
pprint(Xy)

np.save("ant_bee_img", Xy)
```

```python
# 배열을 각각의 배열로 저장하기
pprint(X_pix)
pprint(y_lab)

np.savez('ant_bee_arr', X_pix, y_lab)
```

### numpy.dot() : Dot product

```python
arr1 = np.array([1, 3, 5])
arr2 = np.array([2, 4, 6])

np.dot(arr1, arr2)
```

```python
arr1 = np.array([1, 3, 5])
arr2 = np.array([2, 4, 6, 2])

np.dot(arr1, arr2)  # 에러 발생
```

```python
# dot((2, 2), (2, 2)) => (2, 2)
arr1 = np.array([[1, 3],
                 [2, 4]])

arr2 = np.array([[2, 2], 
                 [1, 3]])

print(arr1.shape)
print(arr2.shape)

np.dot(arr1, arr2)
```

```python
# n차원 * 1차원 : (2, 2)*(2, )  => (2, )
arr1 = np.array([[1, 3],
                 [2, 4]])

arr2 = np.array([2, 2])

print(arr1.shape)
print(arr2.shape)

np.dot(arr1, arr2)
```

```python
# 3차원 * 2차원
arr1 = np.array([[[2, 2], 
                  [1, 1]], 
                 [[1, 2], 
                  [2, 1]]])
arr2 = np.array([[1, 3],
                 [2, 4]])



print(arr1.shape)
print(arr2.shape)

np.dot(arr1, arr2)
np.dot(arr1, arr2).shape
```

```python
arr1 = np.array([[[2, 2], 
                  [1, 1]], 
                 [[1, 2], 
                  [2, 1]]])
arr2 = np.array([[1, 3],
                 [2, 4]])

arr3 = np.array([[1, 2, 3],
                 [2, 3, 1]])


print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

print(np.dot(np.dot(arr1, arr2), arr3).shape)
print(np.dot(np.dot(arr1, arr2), arr3))
```
