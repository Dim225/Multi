###time 모듈
- 참조 : https://wikidocs.net/33/33#time
- time.time() : 현재 시간을 실수 형태로 출력
- time.localtime() : 년, 월, 일, 시, 분, 초 .. 출력

```python
import time

stime = time.time()   # 코드 실행 시작 시간
print(time.time())
print(time.localtime(time.time()))
print(time.strftime("%x", time.localtime(time.time())))

tot = 0
for a in range(10):
    time.sleep(1)      # 1초 대기
    tot += tot + a

print(tot)

print(time.time() - stime)         # 작업 처리 속도 (코드 종료 시간 - 코드 시작 시간)
```

while문
- 무한 반복문: 조건식이 True이면 계속해서 실행
image.png

```python
# 폴더 생성하기

folder_name = "data_mk"

fnames = os.listdir()      # 폴더에 있는 파일 및 하위폴더 가져오기

cnt = 0

while True:
    if folder_name in fnames:      # 폴더명이 존재하는지 확인
        cnt += 1
        folder_name = folder_name[:7]+str(cnt)    # 기본 폴더명에 번호 연결
    else:
        os.mkdir(os.getcwd()+"/0308/"+folder_name)    # 폴더명으로 새로운 폴더 생성
        break       # 반복문 강제 종료

print(os.listdir(os.getcwd()+"/0308/"))
```
