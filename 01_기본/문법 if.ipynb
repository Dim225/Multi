```python
# os 모듈 연결

import os

# print(os.listdir("./data/images/"))
fnames = os.listdir("./data/images/")
print(fnames)
```

```python
import pandas as pd
```

```python
# 데이터 읽어와 정보 저장하기

from PIL import Image

imgInfo = []    # 파일 정보 저장 리스트 변수 생성

for fname in fnames:
    #print("./drive/MyDrive/multi/data/images/"+fname)
    tmp = Image.open("./data/images/"+fname)
    tmpsize = tmp.size    # 이미지 사이즈 튜플 형식으로 저장
    # 파일정보 변수에 저장
    imgInfo.append([fname, tmpsize[0], tmpsize[1]])

    # 강아지 / 고양이 사이즈 변경(강아지_300*300, 고양이_250*250)
    if fname[:3] == "dog":
        tmpResize = tmp.resize((300,300))
    else:
        tmpResize = tmp.resize((250,250)) # 변경 이미지 저장
    tmpResize.save("./data/"+fname[:3]+"/"+fname)


print(imgInfo)
print("=== END ===")
```

###[미션]
- 리스트형 데이터에서 고유값 찾기
- 데이터 : ["사과", "배", "사과", "바나나", "귤", "바나나", "사과", "배", "바나나", "메론"]
- 주어진 데이터에서 고유값을 찾아 리스트형으로 출력하세요

[for/if를 사용]
1. 리스트 생성
2. for문으로 데이터 순차적으로 가져오기
3. 가져온 데이터가 생성한 리스트에 있는지 확인
    * 만약 고유값 리스트 변수의 길이가 0이면 추가
    * 아니면 확인 후 없으면 추가

```python
indata = ["사과", "배", "사과", "바나나", "귤", "바나나", "사과", "배", "바나나", "메론"]

prnlst = []   # 빈 리스트 변수 생성

for a in indata:
    if len(prnlst) == 0:
        prnlst.append(a)
    else:
        cnt = 0        # 입력한 데이터의 갯수를 확인하기 위한 변수
        for b in range(len(prnlst)):
            if a == prnlst[b]:
                cnt = cnt + 1   # cnt += 1
        if cnt == 0:
            prnlst.append(a) 

print(prnlst)
```

[연산자 in, not in 사용]
- a in 데이터 : a 가 데이터에 포함되어 있으면 True / 아니면 False
- a not in 데이터 : a가 데이터에 없으면 True / 있으면 Flase

```python
indata = ["사과", "배", "사과", "바나나", "귤", "바나나", "사과", "배", "바나나", "메론"]

prnlst = []   # 빈 리스트 변수 생성

for a in indata:
    if len(prnlst) == 0:
        prnlst.append(a)
    else:
        if a not in prnlst:
            prnlst.append(a)

print(prnlst)
```

### Text 파일에서 사용자가 입력한 단어의 갯수 구하기
- count()함수 사용 금지
- for과 if를 이용해 count() 함수와 동일한 결과 출력

```python
text1="여러분이 그 여정의 주인공입니다. 여러분이 가는 길이 대학의 역사가 되고 여러분이 개척하는 길이 대한민국의 길이 될 것입니다."

# "길" 이라는 단어 갯수 구하기

cnt = 0    # 갯수 구하는 변수 생성, 초기값으로 0 세팅

for a in text1:
    if a == "길":
        cnt += 1

print(cnt)
```

###사용자로부터 입력 받은 단어 갯수 구하기


```python
text1="여러분이 그 여정의 주인공입니다. 여러분이 가는 길이 대학의 역사가 되고 여러분이 개척하는 길이 대한민국의 길이 될 것입니다."

a = input('단어를 입력해주세요:')
cnt = 0

for i in range(len(text1)):
    if a == text1[i:i + len(a)]:
            cnt += 1
print(cnt)
```

###split() 함수 코드 만들기

```python
# 공백으로 나누어 출력하기

print(text1.split(" "))
```

```python
# for / if 를 이용해 구현

nlst = []   # 단어 저장 변수 생성

tmp = " "    # 구분 단어

word = ""

for a in text1:
    if a != tmp:
        word += a
    else:
        nlst.append(word)
        word = ""

if len(word) != 0:
    nlst.append(word)

print(nlst)
```
