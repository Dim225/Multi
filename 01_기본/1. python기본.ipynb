```python
int1=123
int2=12.3
float1=12.

# type(): 데이터 자료형을 출력
print(type(int1))
print(type(int2))
print(type(int1+float1), int1+float1)

print(int1+int2)  # 숫자형 자료의 + 기호 : 산술연산 + (더하기)
```

```python
str1="문자 숫자 리스트 문자열 텍스트 정수형"
# count('단어') : 문자열에서 단어가 몇개 있는지 확인
print(str1.count("문자"))

# find("문자") : 문자열에서 찾는 문자의 위치 출력
print(str1.find("문자"))

str1.find("문자", 3)
```

```python
# split() : 문자열 나누기
spstr1=str1.split(" ")
print(spstr1)

# join() : 문자열 삽입, 문자사이에 특정 기호등을 삽입할때 사용
str2="12345"
print("-".join(spstr1))
print("-".join(str2))

# len() : 데이터 길이
print(len(str1))

# replace("대상문자", "변경문자") : 문자열 바꾸기
print(str1.replace("문자", "숫자"))
print(str1)
```

#### 리스트형
- 복합형 데이터 구조
- 입력시 대괄호([ ])로 묶어서 입력
- 리스트변수=[값1, 값2, 값3, ... ]
- 인수(값)로는 문자, 숫자, 리스트, 듀퓰, 틱셔너리 등 다양 데이터 형식을 가질수 있음

- 인덱싱/슬라이싱

```python
lst1= [1, 4, 3, 2, 5]
lst2 = [10, 20, "멀티캠퍼스", "파이썬", [100, 200, 300]]
lst3 = ["이순신", "홍길동", "서정동", "김보연", "민창기"]

print(type(lst1))
print(type(lst2))

# 인덱싱/슬라이싱
print(lst2[2])
print(lst2[0:2]) # print(lst2[:2])
print(lst2[-1])
print(lst2[-1][1])  # 2차원 구조의 데이터 출력

# 리스트형 데이터 수정
lst2[3] = "python"
print(lst2)

# append() : 리스트의 끝에 값 추가
lst2.append("서정동")
print(lst2)

lst1.sort()  # 오름차순 정렬
print(lst1)

lst3.sort()
print(lst3)

# 객체.insert(위치, 값) : 중간 삽입
lst1.insert(0, 6)
lst1

# 개체 삭제
del lst2
del lst3[2]

lst3
```

#### 딕셔너리형
- 복합 자료형
- 입력시 중괄호({ })로 묶어서 입력
- 딕셔너리변수={"key1":값1, "key2":값2, ....}
- key: 대표 이름
- 값: key가 가리키는 값, 숫자/문자/리스트/딕셔너러 등


```python
dict1 = {"번호":1, "이름":"홍길동", "성별":"남", "국어":100}
dict2 = {"번호":[1, 2, 3, 4, 5],
         "이름":["AAA", "BBB", "CCC", "DDD", "EEE"],
         "성별":["남", "여", "남", "여", "남"],
         "국어":[100, 95, 85, 75, 80]}

print(dict1.keys())    # key 이름 출력
print(dict1.values())  # 값 전체 출력
print()
print(dict2.keys())
print(dict2.values())

print(dict2["이름"])    # 특정 키에 해당하는 값만 출력
```

```python
# pandas를 이용한 데이터프레임 구조 변경

df1=pd.DataFrame(dict2)
df1
```
